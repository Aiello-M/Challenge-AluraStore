## <font color=lightblue> **2. Análisis del conjunto de datos**

### 2.1 Análisis de facturación

Se calcula la facturación total de cada tienda a partir del precio de los productos vendidos <br>y se comparan los resultados mediante un gráfico de torta.

```python
def facturacion_total(df: pd.DataFrame) -> float:
    '''
    Función para calcular la facturación total de una tienda a partir de su DataFrame de ventas.

    Parámetros
    df : pandas.DataFrame
        DataFrame que contiene los productos vendidos por la tienda.
        Incluye una columna llamada 'Precio'.

    Return
    float
        Facturación total de la tienda.
    '''
    # Suma total de los precios de los productos vendidos
    return df['Precio'].sum()
```

```python
# Calcular de la facturacion total de cada tienda

facturacion_tienda1 = facturacion_total(tienda)
facturacion_tienda2 = facturacion_total(tienda2)
facturacion_tienda3 = facturacion_total(tienda3)
facturacion_tienda4 = facturacion_total(tienda4)
```

```python
# DataFrame con la facturación de cada tienda

df_facturaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación': [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]
})

df_facturaciones
```

```python
# Mostrar los resultados: Facturación por tienda

print('Facturación por tienda\n')
print(f"Tienda 1: ${facturacion_tienda1:,.0f}")
print(f"Tienda 2: ${facturacion_tienda2:,.0f}")
print(f"Tienda 3: ${facturacion_tienda3:,.0f}")
print(f"Tienda 4: ${facturacion_tienda4:,.0f}")
```

```python
# Gráfico de torta: Facturación total por tienda

plt.figure(figsize=(6, 6))
colores = ['#0964c2', '#1e7ade', '#6baed6', '#bdd7e7']

plt.pie(
    df_facturaciones['Facturación'],
    labels=df_facturaciones['Tienda'],
    autopct=lambda pct: f'{pct:.1f}%\n(${pct/100 * df_facturaciones["Facturación"].sum()/1e9:.2f}B)',
    startangle=90,
    colors=colores,
    explode=(0, 0, 0, 0.05),
    textprops={'fontsize': 11, 'color': '#333333'}
)

plt.suptitle('Facturación total por tienda', fontsize=17)
plt.title('La Tienda 4 presenta la menor facturación total', fontsize=14, color='#282a2c', pad=10)

plt.tight_layout()
plt.show()
```

```python
# Identificar la tienda con menor facturación
tienda_menor_facturacion = df_facturaciones.sort_values('Facturación').iloc[0]
print(f"\nTienda con menor facturación: {tienda_menor_facturacion['Tienda']} con ${tienda_menor_facturacion['Facturación']:,.0f}")
```

```python
# Análisis complementario
print("Análisis complementario\n")
print(f"* Volumen total: Tienda 1={len(tienda)}, Tienda 2={len(tienda2)}, Tienda 3={len(tienda3)}, Tienda 4={len(tienda4)}")

# pecio promedio por transacción
precio_prom_t1 = facturacion_tienda1 / len(tienda)
precio_prom_t2 = facturacion_tienda2 / len(tienda2)
precio_prom_t3 = facturacion_tienda3 / len(tienda3)
precio_prom_t4 = facturacion_tienda4 / len(tienda4)

print(f"\n* Precio promedio por transacción:")
print(f"  -> Tienda 1: ${precio_prom_t1:,.0f}")
print(f"  -> Tienda 2: ${precio_prom_t2:,.0f}")
print(f"  -> Tienda 3: ${precio_prom_t3:,.0f}")
print(f"  -> Tienda 4: ${precio_prom_t4:,.0f}")

# diferencia relativa
diferencia_relativa = ((facturacion_tienda1 - facturacion_tienda4) / facturacion_tienda4) * 100
print(f"\n* Tienda 4 factura {diferencia_relativa:.1f}% menos que la Tienda 1")
```

#### Observaciones

Las cuatro tiendas presentan facturaciones relativamente similares, con porcentajes entre 24% y 27% del total, y volúmenes de venta prácticamente idénticos (aproximadamente 2,359 transacciones cada una).

**La Tienda 4 registra la menor facturación** con \$1,038,375,700 (24.2% del total), mientras que la **Tienda 1** lidera con \$1,150,880,400 (26.8%). La diferencia entre la tienda con mayor y menor facturación es de \$112.5 millones, lo que representa un 10.8% menos de ingresos para la Tienda 4.

### 2.2 Ventas por categoría

Se calcula la cantidad de productos vendidos por categoría para cada tienda, mostrando los resultados mediante un gráfico de barras agrupadas.


```python
# Verificar las categorías de productos existentes
tienda['Categoría del Producto'].unique()
```

```python
def ventas_por_categoria(df: pd.DataFrame) -> pd.Series:
    '''
    Función para calcular la cantidad de ventas por categoría en una tienda.

    Parámetros
    df : pandas.DataFrame
        DataFrame que contiene los productos vendidos por la tienda.
        Incluye una columna llamada 'Categoría del Producto'.

    Return
    pandas.Series
        Serie con la cantidad de ventas por categoría.
        El índice representa las categorías y los valores la cantidad de ventas.
    '''
    return df['Categoría del Producto'].value_counts().rename('Ventas')
```

```python
# Calcular la cantidad de ventas por categoria
ventas_categ_tienda1 = ventas_por_categoria(tienda)
ventas_categ_tienda2 = ventas_por_categoria(tienda2)
ventas_categ_tienda3 = ventas_por_categoria(tienda3)
ventas_categ_tienda4 = ventas_por_categoria(tienda4)
```

### 2.3 Calificación promedio de la tienda

Se calcula la calificación promedio de cada tienda, se comparan los resultados mediante un gráfico de barras  horizontal y se identifica la tienda con la menor calificación promedio.

```python
def promedio_calificacion(df: pd.DataFrame) -> float:
    '''
    Función para calcular la calificación promedio de una tienda.

    Parámetros
    df : pandas.DataFrame
        DataFrame que contiene las ventas de una tienda.
        Incluye una columna llamada 'Calificación'.

    Return
    float
        Calificación promedio de los clientes.
    '''
    return df['Calificación'].mean()
```

```python
# Calcular la calificación promedio de cada tienda

calificacion_tienda1 = promedio_calificacion(tienda)
calificacion_tienda2 = promedio_calificacion(tienda2)
calificacion_tienda3 = promedio_calificacion(tienda3)
calificacion_tienda4 = promedio_calificacion(tienda4)
```

```python
# DataFrame con las calificaciones promedio
df_calificacion = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación promedio': [
        round(calificacion_tienda1, 2),
        round(calificacion_tienda2, 2),
        round(calificacion_tienda3, 2),
        round(calificacion_tienda4, 2)
    ]
})

df_calificacion
```

```python
# Mostrar los resultados: Calificación promedio
print('Calificación promedio por tienda (sobre 5)\n')
print(f"Tienda 1: {calificacion_tienda1:.2f}")
print(f"Tienda 2: {calificacion_tienda2:.2f}")
print(f"Tienda 3: {calificacion_tienda3:.2f}")
print(f"Tienda 4: {calificacion_tienda4:.2f}")
```

```python
# Gráfico de barras horizontales: Calificación promedio de clientes por tienda

# Ordenar de menor a mayor para que la peor quede arriba
df_calif_ordenado = df_calificacion.sort_values('Calificación promedio', ascending=True)

plt.figure(figsize=(8, 5))
colores = ['#bdd7e7', '#6baed6', '#1e7ade', '#0964c2']

barras = plt.barh(
    df_calif_ordenado['Tienda'],
    df_calif_ordenado['Calificación promedio'],
    color=colores,
    height=0.6
)

plt.suptitle('Calificación promedio por tienda', fontsize=17)
plt.title('Las Tiendas 1 y 4 registran menor satisfacción del cliente', fontsize=14, color='#282a2c', pad=15)
plt.xlabel('Calificación promedio (sobre 5)', fontsize=11, labelpad=20)
plt.xlim(0, 5)

# Valores al final de cada barra
for barra in barras:
    ancho = barra.get_width()
    plt.text(
        ancho + 0.05,
        barra.get_y() + barra.get_height() / 2,
        f'{ancho:.2f}',
        va='center',
        ha='left',
        fontsize=10,
        color='#333333'    # Mismo estilo que el Gráfico 5
    )

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# plt.margins(y=0.15) lo pongo en eje x? x= 0.15 o no lo pongo?
plt.tight_layout()
plt.show()
```

### 2.4 Productos más y menos vendidos

Se identifican los productos con mayor y menor cantidad de ventas en cada tienda, comparando los resultados mediante gráficos de barras.

```python
# Verificar los productos únicos existentes en las Tiendas

print(f"Total de productos únicos: {tienda['Producto'].nunique()}")

print(f"\nEjemplos de productos:\n{tienda['Producto'].unique()[:10]}")
```

```python
def ventas_por_producto(df: pd.DataFrame) -> pd.Series:
  '''
  Función para calcular la cantidad de ventas por producto en una tienda.

  Parámetros
  df : pandas.DataFrame
      DataFrame que contiene los productos vendidos por la tienda.
      Incluye una columna llamada  'Producto'.

  Return
  pandas.Series
      Serie con la cantidad de ventas por producto en una Tienda.
      El índice representa los productos y los valores la cantidad de ventas.
  '''
  return df['Producto'].value_counts()
```

```python
# Calcular las ventas de productos por tienda
ventas_producto_tienda1 = ventas_por_producto(tienda)
ventas_producto_tienda2 = ventas_por_producto(tienda2)
ventas_producto_tienda3 = ventas_por_producto(tienda3)
ventas_producto_tienda4 = ventas_por_producto(tienda4)
```

```python
# Diccionario con las ventas de productos por tienda
ventas_por_tienda = {
    'Tienda 1': ventas_producto_tienda1,
    'Tienda 2': ventas_producto_tienda2,
    'Tienda 3': ventas_producto_tienda3,
    'Tienda 4': ventas_producto_tienda4
}
```

```python
# Mostrar resultados: productos más y menos vendidos por tienda
print('Top 3 productos más y menos vendidos por tienda\n')

for nombre_tienda, ventas in ventas_por_tienda.items():
    top_3 = ventas.head(3)
    bottom_3 = ventas.tail(3)

    print(f'{nombre_tienda}:')
    print(f"  Más vendidos:")
    for i, (prod, cant) in enumerate(top_3.items(), 1):
        print(f"    {i}. {prod} ({cant} unidades)")

    print(f"  Menos vendidos:")
    for i, (prod, cant) in enumerate(bottom_3.items(), 1):
        print(f"    {i}. {prod} ({cant} unidades)")
    print()
```

```python
# Gráfico de barras: Top 3 más y menos vendidos por tienda

max_venta_global = max(ventas.max() for ventas in ventas_por_tienda.values())
limite_y = max_venta_global * 1.08

plt.figure(figsize=(13, 12))
plt.suptitle('Top 3 productos más y menos vendidos por tiendaa', fontsize=18)
azules = ['#084594', '#2171b5', '#4292c6'] # de mas oscuro (más vendido) a mas claro
rojos = ['#fcbba1', '#fc9272', '#ef3b2c']  # de mas claro a mas oscuro (menos vendido)

for i, (nombre_tienda, ventas) in enumerate(ventas_por_tienda.items()):
    top_3 = ventas.head(3)
    bottom_3 = ventas.tail(3)

    # Combinar datos para el gráfico
    productos_plot = list(top_3.index) + list(bottom_3.index)
    valores_plot = list(top_3.values) + list(bottom_3.values)
    colores_plot = azules + rojos

    ax = plt.subplot(2, 2, i + 1)
    x_pos = range(len(productos_plot))
    barras = plt.bar(x_pos, valores_plot, color=colores_plot, width=0.7)

    # subgráfico
    plt.title(nombre_tienda, pad=15, fontsize=16)
    plt.ylim(0, limite_y)
    plt.xticks(x_pos, productos_plot, rotation=45, ha='right', fontsize=9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    if i % 2 == 0:
        plt.ylabel('Cantidad vendida', fontsize=11, labelpad=20)

    # etiqueta de datos
    for barra in barras:
        altura = barra.get_height()
        plt.text(
            barra.get_x() + barra.get_width() / 2,
            altura + (limite_y * 0.01),
            f'{altura:.0f}',
            ha='center', va='bottom', fontweight='bold', color='#333333'
        )


plt.tight_layout(rect=[0, 0.03, 1, 0.94], h_pad=6.0, w_pad=3.0)
plt.show()
```

```python
# ANÁLISIS COMPLEMENTARIO: comprobación de precios de los productos más vendidos

print("Precios del Top 3 de productos vendidos:")

# Top 3 de la Tienda 1 (Mayor facturación)
top3_t1 = ventas_por_tienda['Tienda 1'].head(3).index
print("\nTienda 1 (Mayor facturación global):")
for prod in top3_t1:
    precio = tienda[tienda['Producto'] == prod]['Precio'].iloc[0]
    print(f"  * {prod}: ${precio:,.2f}")


# Top 3 de la Tienda 4 (Menor facturación)
top3_t4 = ventas_por_tienda['Tienda 4'].head(3).index
print("\nTienda 4 (Menor facturación global):")
for prod in top3_t4:
    precio = tienda4[tienda4['Producto'] == prod]['Precio'].iloc[0]
    print(f"  * {prod}: ${precio:,.2f}")

print("\n__________________________________________\n")
print("Precios del Bottom 3 de productos vendidos:")

# Bottom 3 de la Tienda 1 (Mayor facturación)
bottom3_t1 = ventas_por_tienda['Tienda 1'].tail(3).index
print("\nTienda 1 (Mayor facturación global):")
for prod in bottom3_t1:
    precio = tienda[tienda['Producto'] == prod]['Precio'].iloc[-1]
    print(f"  * {prod}: ${precio:,.2f}")


# Bottom 3 de la Tienda 4 (Menor facturación)
bottom3_t4 = ventas_por_tienda['Tienda 4'].tail(3).index
print("\nTienda 4 (Menor facturación global):")
for prod in bottom3_t4:
    precio = tienda4[tienda4['Producto'] == prod]['Precio'].iloc[-1]
    print(f"  * {prod}: ${precio:,.2f}")
```

#### Observaciones

No existe un único producto que esté dentro del top más vendido de todas las tiendas. Cada tienda tiene su propio top 3, sin una tendencia clara común entre las sucursales.

Todas las tiendas registran cantidades similares de productos más vendidos, oscilando entre 57 y 65 unidades en el top 3, sin diferencias significativas. Los productos menos vendidos también presentan volúmenes comparables (32-36 unidades) en todas las tiendas.

### 2.5 Envío promedio por tienda

Se calcula el costo promedio de envío de cada tienda y se comparan los resultados mediante un gráfico de barras.

```python
def promedio_envio(df: pd.DataFrame) -> float:
  '''
    Función para calcular el costo promedio de envío de una tienda.

    Parámetros
    df : pandas.DataFrame
        DataFrame que contiene las ventas de la tienda.
        Incluye una columna llamada 'Envío'.

    Return
    float
        Costo promedio de envío de la tienda.
    '''
  return df['Costo de envío'].mean()
```

```python
# Calcular el costo promedio de envío de cada tienda
promedio_envio_tienda1 = promedio_envio(tienda)
promedio_envio_tienda2 = promedio_envio(tienda2)
promedio_envio_tienda3 = promedio_envio(tienda3)
promedio_envio_tienda4 = promedio_envio(tienda4)
```
